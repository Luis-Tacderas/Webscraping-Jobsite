```python
import time
import logging
import logging.config
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
import pandas as pd
start = time.time()
```

```python
# set up logger tracking and information 
logger = logging.getLogger(__name__)
logger.setLevel(logging.INFO)
ch = logging.StreamHandler()
ch.setLevel(logging.INFO)
logger.addHandler(ch)
formatter = logging.Formatter(
    '%(asctime)s %(levelname)s %(lineno)d\
    :%(filename)s(%(process)d) - %(message)s')
ch.setFormatter(formatter)

logging.getLogger('selenium').setLevel(logging.CRITICAL)
```

```python
# open up a browser to start testing
def get_browser():
    logger.info('Configuring browser')
    chrome_options = wd.ChromeOptions()
    chrome_options.add_argument('log-level=3')
    browser = wd.Chrome(options=chrome_options)
    return browser

browser = get_browser()
```

```python
# Open a new Chrome window
#url = 'https://www.kaiserpermanentejobs.org/search-jobs'

url = 'https://www.kaiserpermanentejobs.org/search-jobs/Data%20Reporting%20and%20Analytics%20Consultant/Pleasanton%2C%20CA/641-12921/1/4/6252001-5332921-5322745-5383777/37x66243/-121x87468/50/2'
browser.get(url)
    
time.sleep(1)
```

```python
# Search List
#https://stackoverflow.com/questions/47436151/how-to-loop-through-only-li-elements-inside-a-ul
resultSet = browser.find_element(By.XPATH,"//*[@id='search-results-list']/ul")
```

```python
date_posted = resultSet.find_elements(By.XPATH, '//span[@class = "job-date-posted"]')
date_posted_list = []
for item in date_posted:  
    date_posted_list.append(item.text)
date_posted_df = pd.DataFrame(date_posted_list)
```

```python
loc = resultSet.find_elements(By.XPATH, '//span[@class = "job-location"]')
location_list = []
for item in loc:  
    location_list.append(item.text)
location_df = pd.DataFrame(location_list)
```

```python
job_item = resultSet.find_elements(By.TAG_NAME,"h2")
```

```python
job_list = []
for item in job_item:
    job_list.append(item.text)
job_list_df = pd.DataFrame(job_list)
```
